```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/monty_hall_sample.csv')
df
```

```python
df.info()
```

```python
# Number of wins for each strategy
df[df['won'] == 1].groupby('strategy').count()['won'].plot(kind='bar')
plt.xlabel('Strategy')
plt.ylabel('Number of Wins')
plt.title('Monty Hall Winning Results')
plt.xticks(rotation=0) # Changes the text orientation of the x-axis labels to horizontal
plt.show()
```

```python
# Number of loses for each strategy
df[df['won'] == 0].groupby('strategy').count()['won'].plot(kind='bar')
plt.xlabel('Strategy')
plt.ylabel('Number of Loses')
plt.title('Monty Hall Loosing Results')
plt.xticks(rotation=0) # Changes the text orientation of the x-axis labels to horizontal
plt.show()
```
